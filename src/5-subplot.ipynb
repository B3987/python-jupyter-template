### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0, 10, 100)
y_line = np.sin(x)
x_bar = ['A', 'B', 'C', 'D']
y_bar = [5, 10, 8, 12]
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
labels_pie = ['One', 'Two', 'Three', 'Four']
sizes_pie = [15, 30, 45, 10]
colors_pie = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode_pie = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Two')


fig, axs = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Subplot Layout')


axs[0, 0].plot(x, y_line)
axs[0, 0].set_title('Line Chart')


axs[0, 1].bar(x_bar, y_bar)
axs[0, 1].set_title('Bar Chart')


axs[1, 0].scatter(x_scatter, y_scatter)
axs[1, 0].set_title('Scatter Plot')


axs[1, 1].pie(sizes_pie, explode=explode_pie, labels=labels_pie, colors=colors_pie,
        autopct='%1.1f%%', shadow=True, startangle=90)
axs[1, 1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
axs[1, 1].set_title('Pie Chart')


plt.tight_layout(rect=[0, 0.03, 1, 0.95]) 


plt.show()
```
